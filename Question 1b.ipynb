```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = 3
index = np.arange(x)
width = 0.25

years = ['2018', '2019', '2020']
companies = ['Amazon', 'Facebook', 'Apple']

amazon_revenue = [280, 386, 469]
amazon_bar = plt.bar(index, amazon_revenue, width)

facebook_revenue = [55, 70, 85]
facebook_bar = plt.bar(index + width, facebook_revenue, width)

apple_revenue = [265, 229, 274]
apple_bar = plt.bar(index + (width * 2), apple_revenue, width)

plt.xlabel('Year')
plt.xticks(index + width, years) 
plt.ylabel('Revenue')
plt.title('Yearly Revenue by Company')
plt.legend(companies)
plt.show()
```
