**TASK 5** `Pclass` `15 Minutes`
- Visualize Passenger Class Distribution using two ways (Pclass (Ticket class): A proxy for socio-economic status, with 1 being the highest class and 3 the lowest.)

**TASK 5.1** 
- Get the unique values of Pclass column

```python
# Get the Unique Values for `Pclass`
df['Pclass'].unique()
```

**TASK 5.2**
- Get the counts values of each class in Pclass

```python
# Get the counts of each value of `Pclass`
class_counts = df['Pclass'].value_counts()
class_counts
```

Plot the counts of `Pclass` in **First Way**

```python
# Plot the value Count of `Pclass` in a bar chart
class_counts.plot(kind = 'bar')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
```

Plot the counts of `Pclass` in **Second Way**

```python
# Plot the value Count of `Pclass` in a pie chart with Percentage values on it
class_counts.plot(kind = 'pie', autopct = '%1.1f%%')
plt.title('Class Distribution')
```

**TASK 5.3**
- map values of Pclass from `{0,1,2}` to `{1st_class, 2nd_class, 3rd_class}` by creating a new column `Pclass_categorical`

```python
# Create a column `Pclass_Categorical` with the new mapping values of Pclass
df['Pclass_categorical'] = df['Pclass'].map({1:"1st_class", 2: "2nd_class", 3: "3rd_class"})
df[['Pclass', 'Pclass_categorical']]
Pclass_categorical = df['Pclass_categorical'].value_counts()
```

**TASK 5.4**
- plot `Pclass_Categorical` with bar chart using the following colors ['blue','green','red']
- With writing a count on each bar

```python
#Plot `Pclass_Categorical` in a bar chart with a specific color
# Check the discussion for Hint 
# It should be similar to the blew plot
ax = Pclass_categorical.plot(kind = 'bar', color = ['blue', 'green', 'red'])
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.ylabel('Count')
plt.title('Class Distribution')
```
