```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Year': [2016, 2017, 2018, 2019, 2020, 2021],
    'Revenue': [135, 177, 232, 280, 386, 469]
}

df = pd.DataFrame(data)

slope = 67
y_intercept = -134960

df['Predicted'] = (df['Year'] * slope) + y_intercept

df['Detrended'] = df['Revenue'] - df['Predicted']

sns.lineplot(data = df, x = 'Year', y = 'Detrended')
plt.axhline(y = 0.0, color = 'Red', linestyle = '-')
plt.xlabel('Year')
plt.ylabel('Detrended Revenue')
plt.title("Amazon's Yearly Revenue From 2016 To 2021 - Detrended")
plt.show()
```
